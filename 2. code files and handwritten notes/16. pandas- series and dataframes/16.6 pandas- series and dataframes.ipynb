### 2. DataFrame Operations

Create a Pandas DataFrame with 3 columns (Col1, Col2, Col3) and 4 rows filled with random integers between 1 and 99.
Then add a new column called Product that stores the product of Col1 and Col2.

```python
import pandas as pd
import numpy as np

# -------------------------------------------------------------
# This block demonstrates creating and manipulating a Pandas DataFrame.
#
# 1. We generate a 5x3 DataFrame with random integers (1–99) to simulate tabular data.
# 2. Columns are named 'Col1', 'Col2', 'Col3'.
# 3. We then create a new column 'Product' which multiplies the first two columns.
#    This shows how to perform vectorized operations across columns.
# 4. Using vectorized operations (df['Col1'] * df['Col2']) is faster and cleaner than looping.
# -------------------------------------------------------------

# Create a 4x3 DataFrame with random integers
df_table = pd.DataFrame(
    np.random.randint(1, 30, size=(4, 3)),
    columns=['Col1', 'Col2', 'Col3']
)

print("Original DataFrame:")
print(df_table)

# Add a new column 'Product' = Col1 * Col2
df_table['Product'] = df_table['Col1'] * df_table['Col2']

print("\nDataFrame with 'Product' column added:")
print(df_table)
```

Create a Pandas DataFrame with 6 columns (A-F) and 4 rows filled with random integers between 20 and 40.
Compute the sum of each row and the sum of each column.

```python
import pandas as pd
import numpy as np

# -------------------------------------------------------------
# This block demonstrates creating a Pandas DataFrame and performing aggregation operations.
#
#  To compute the sum of each row: df.sum(axis=1)
#    - axis=1 means we move across columns (horizontal direction)
#  To compute the sum of each column: df.sum(axis=0)
#    - axis=0 means we move down rows (vertical direction)
#  This illustrates how to quickly get row-wise and column-wise totals without using loops.
# -------------------------------------------------------------

# Create a 4x6 DataFrame with random integers
df_table = pd.DataFrame(
    np.random.randint(20, 40, size=(4, 6)),
    columns=['A', 'B', 'C', 'D', 'E', 'F']
)

print("Original DataFrame:")
print(df_table)

# Compute row-wise sum (sum across columns)
row_sums = df_table.sum(axis=1)

# Compute column-wise sum (sum down rows)
column_sums = df_table.sum(axis=0)

print("\nRow-wise sum:")
print(row_sums)
print("\nColumn-wise sum:")
print(column_sums)
```

### 3. Data Cleaning

Create a Pandas DataFrame with 5 rows and columns. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.

```python
'''
  NaN (Not a Number) values are intentionally inserted to simulate missing data. Pandas represents missing numerical
    values using NaN from NumPy.

  `df.iloc[row_index, column_index]` is used to access elements by **integer position** (row number and column number).
    Example: df.iloc[0,1] → first row, second column.

    `df.loc[]` accesses data by **row label and column label** instead of position.
    Example: df.loc[0, 'B'] → row label 0, column 'B'
    Since this DataFrame uses default row labels (0,1,2,...), `.loc[0, 'B']` and `.iloc[0, 1]` refer to the same cell.

    In contrast, `.loc[row_label, column_label]` accesses data using **labels** instead of numeric positions.

  Missing values are handled using `fillna()`.
    `df.fillna(df.mean(), inplace=True)` replaces NaN values with the **mean of their respective columns**.

  `inplace=True` modifies the original DataFrame directly instead of creating a new one.

 This technique is commonly used in data preprocessing to handle missing data before performing analysis or modeling.
-------------------------------------------------------------'''
df = pd.DataFrame(np.random.randint(1, 30, size=(5, 5)), columns=['A', 'B', 'C','D','E'])
print("Original DataFrame:")
print(df)

# Introduce some NaN values
df.iloc[0, 1] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[4, 0] = np.nan
print("DataFrame with NaN values:")
print(df)

# Fill the NaN values with the mean of the respective columns
df.fillna(df.mean(), inplace=True)
print("DataFrame with NaN values filled:")
print(df)
```

### 4. Data Aggregation

Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.

```python
'''
Creating a DataFrame using a dictionary

In Pandas, a dictionary can be used to construct a DataFrame where:
    • dictionary keys   → column names
    • dictionary values → column data

In this example:
    'Category' : np.random.choice(['A','B','C'], size=10)
        - randomly selects 10 elements from the list ['A','B','C']
        - this simulates categorical data

    'Value' : np.random.randint(1,100, size=10)
        - generates 10 random integers between 1 and 99
        - this represents numerical data

Both arrays have the same length (10), which determines the number
of rows in the DataFrame.

Pandas combines these arrays row-wise to form a table like:

    Category   Value
    A          45
    B          12
    C          77
    ...

After the DataFrame is created:
    df.groupby('Category')['Value'].agg(['sum','mean'])

1. groupby('Category')
   groups rows based on unique values in the Category column.

2. ['Value']
   selects the Value column for aggregation.

3. agg(['sum','mean'])
   calculates multiple statistics for each group:
        - sum  → total value per category
        - mean → average value per category
'''

df = pd.DataFrame({'Category': np.random.choice(['A', 'B', 'C'], size=10), 'Value': np.random.randint(1, 100, size=10)})
print("Original DataFrame:")
print(df,'\n')

# Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category
grouped = df.groupby('Category')['Value'].agg(['sum', 'mean'])
print("Grouped DataFrame:")
print(grouped)
```

### 5. Merging DataFrames

Create two Pandas DataFrames with a common column. Merge the DataFrames using the common column.

```python
'''
pd.merge() is used to combine two DataFrames based on one or more common columns (similar to SQL joins).

In this example:
    merged = pd.merge(df1, df2, on='Key')

• on='Key' Specifies the column used as the common key for matching rows between the two DataFrames.

Default parameters of pd.merge():
    how='inner'   → default join type (only matching keys kept)
    on=None       → if not provided, Pandas uses common column names
    sort=False    → result is not automatically sorted
    suffixes=('_x','_y')
                   → used when both DataFrames have columns with
                     the same name (other than the key)

Common join types:
    inner (default) → only rows with matching keys in both tables
    left            → all rows from left DataFrame + matching right
    right           → all rows from right DataFrame + matching left
    outer           → union of keys from both DataFrames
'''

df1 = pd.DataFrame({'Key': ['A', 'B', 'C', 'D'], 'Value1': np.random.randint(1, 10, size=4)})
df2 = pd.DataFrame({'Key': ['A', 'B', 'C', 'E'], 'Value2': np.random.randint(1, 10, size=4)})
print("DataFrame 1:")
print(df1,'\n')
print("DataFrame 2:")
print(df2,'\n')


merged = pd.merge(df1, df2, on='Key')
print("Merged DataFrame:")
print(merged)
```

Create two Pandas DataFrames with different columns. Concatenate the DataFrames along the rows and along the columns.

```python
df1 = pd.DataFrame({'A': np.random.randint(1, 10, size=3), 'B': np.random.randint(1, 10, size=3)})
df2 = pd.DataFrame({'C': np.random.randint(10, 20, size=3), 'D': np.random.randint(10, 20, size=3)})
print("DataFrame 1:")
print(df1)
print("DataFrame 2:")
print(df2,'\n')

'''
pd.concat([df1, df2]) is used to combine multiple DataFrames along a particular axis.

The DataFrames are passed as a list because concat() expects an iterable collection of objects to combine.

Important parameter:axis

axis=0 (default) → concatenate along rows (vertical stacking)
    - rows of df2 are added below rows of df1
    - column names must match to align properly
    - if columns differ, missing values appear as NaN

axis=1 → concatenate along columns (horizontal stacking)
    - columns of df2 are added next to df1
    - rows are aligned based on their index values
'''


# Concatenate the DataFrames along the rows
concat_rows = pd.concat([df1, df2], axis=0)
print("Concatenated DataFrame (rows):")
print(concat_rows,'\n')

# Concatenate the DataFrames along the columns
concat_columns = pd.concat([df1, df2], axis=1)
print("Concatenated DataFrame (columns):")
print(concat_columns)
```

### 6. Time Series Analysis

Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.

```python
'''
1. Creating a date range:
   pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
   generates a sequence of dates from Jan 1, 2022 to Dec 31, 2022.
   The parameter freq='D' means the frequency is daily.

2. Creating the DataFrame:
   The date range becomes a column named 'date'. A second column
   called 'data' is created using random integers between 0 and 99,
   simulating daily measurements or observations.

3. Setting the index:
   df.set_index('date', inplace=True)
   converts the 'date' column into the DataFrame index.
   This is important because many time-series operations in Pandas
   require a datetime index.

4. Resampling:
   df.resample('ME') groups the data by month.
   'ME' represents month-end frequency.

5. Aggregation:
   df.mean() calculates the average value of the 'data' column
   for each month.
'''

date_rng = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0, 100, size=(len(date_rng)))
df.set_index('date', inplace=True)
print("Original DataFrame:")
print(df)

# Resample the DataFrame to compute the monthly mean of the values
monthly_mean = df.resample('ME').mean()
print("Monthly mean DataFrame:")
print(monthly_mean)
```

Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
date_rng = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0, 100, size=(len(date_rng)))
df.set_index('date', inplace=True)
print("Original DataFrame:")
print(df,'\n')

# Compute the rolling mean with a window of 7 days
rolling_mean = df.rolling(window=7).mean()
print("Rolling mean DataFrame:")
print(rolling_mean)

'''
The rolling() function is used to perform calculations over a moving (or sliding) window of values in a DataFrame or Series.

df.rolling(window=7)

• window=7
  Specifies the size of the moving window. Here it means each calculation considers the current row and the previous 6 rows,
  giving a total window of 7 observations.

• .mean()
  Computes the average value within each rolling window.

Since the dataset is indexed by dates (time-series data), this operation effectively calculates a 7-day moving average.

Important notes:
    - The first few rows will contain NaN values because there are not enough previous observations to fill the window.
    - Rolling statistics are commonly used to smooth noisy time-series data and identify trends.

The resulting DataFrame shows the rolling (moving) average of the values over a 7-day window.
'''
```

### 7. MultiIndex DataFrame

Create a Pandas DataFrame with a MultiIndex (hierarchical index). Perform some basic indexing and slicing operations on the MultiIndex DataFrame.

```python
arrays = [['A', 'A', 'B', 'B'], ['one', 'two', 'one', 'two']]
index = pd.MultiIndex.from_arrays(arrays, names=('Category', 'SubCategory'))
df = pd.DataFrame(np.random.randint(1, 100, size=(4, 3)), index=index, columns=['Value1', 'Value2', 'Value3'])
print("MultiIndex DataFrame:")
print(df,'\n')

# Basic indexing and slicing operations
print("Indexing at Category 'A':")
print(df.loc['A'],'\n')

print("Slicing at Category 'B' and SubCategory 'two':")
print(df.loc[('B', 'two')])

'''
MultiIndex (Hierarchical Index) in Pandas

A MultiIndex allows a DataFrame to have multiple levels of indexing on the rows (or columns). 
This is useful for representing higher-dimensional data in a tabular format.

1. Creating the MultiIndex:
   arrays = [['A','A','B','B'], ['one','two','one','two']]

   These two lists represent two levels of indexing:
       Level 1 → Category     : A, B
       Level 2 → SubCategory  : one, two

   pd.MultiIndex.from_arrays(arrays, names=('Category','SubCategory'))
   combines these arrays into a hierarchical index with named levels.

2. Creating the DataFrame:
   The MultiIndex is used as the row index, and three columns
   ('Value1','Value2','Value3') contain random numerical data.

   The DataFrame structure looks conceptually like:

        Category  SubCategory   Value1 Value2 Value3
        A         one
                  two
        B         one
                  two

3. Indexing with MultiIndex:
   df.loc['A']
       retrieves all rows where the first level (Category) is 'A'.

   df.loc[('B','two')]
       accesses the specific row where:
           Category = 'B'
           SubCategory = 'two'

MultiIndex makes it possible to perform complex hierarchical grouping, slicing, and aggregation operations efficiently.
'''
```

Create a Pandas DataFrame with MultiIndex consisting of 'Category' and 'SubCategory'. Fill the DataFrame with random data and compute the sum of values for each 'Category' and 'SubCategory'.

```python

arrays = [['A', 'A', 'B', 'B', 'C', 'C'], ['one', 'two', 'one', 'two', 'one', 'two']]
index = pd.MultiIndex.from_arrays(arrays, names=('Category', 'SubCategory'))
df = pd.DataFrame(np.random.randint(1, 100, size=(6, 3)), index=index, columns=['Value1', 'Value2', 'Value3'])
print("MultiIndex DataFrame:")
print(df,'\n')

sum_values = df.groupby(['Category', 'SubCategory']).sum()
print("Sum of values:")
print(sum_values)
```

### 8. Pivot Tables

Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.

```python
# Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'
date_rng = pd.date_range(start='2022-01-01', end='2022-01-10', freq='D')
df = pd.DataFrame({'Date': np.random.choice(date_rng, size=20), 'Category': np.random.choice(['A', 'B', 'C'], size=20), 'Value': np.random.randint(1, 100, size=20)})
print("Original DataFrame:")
print(df,'\n')

# Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'
pivot_table = df.pivot_table(values='Value', index='Date', columns='Category', aggfunc='sum')
print("Pivot Table:")
print(pivot_table)

'''
Creating a Pivot Table in Pandas

A pivot table is used to summarize and reorganize data by grouping rows and columns based on specific variables.

df.pivot_table(...) restructures the DataFrame so that values from one column are aggregated and displayed across categories.

Parameters used here:
- values='Value'
    Specifies the column whose data will be aggregated.

- index='Date'
    Determines the rows of the pivot table.
    Each unique value in the 'Date' column becomes a row.

- columns='Category'
    Determines the columns of the pivot table.
    Each unique value in the 'Category' column becomes a column.

- aggfunc='sum'
    Defines the aggregation function applied to the values.

Conceptually the operation converts data from a "long format" into a "wide format", making it easier to analyze relationships
between categories across different dates.
Pivot tables are commonly used in data analysis to summarize, group, and compare values across multiple dimensions.
'''
```

Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.

### 9. Applying Functions

Create a Pandas DataFrame. Apply a function that doubles the values of the DataFrame.

```python
df = pd.DataFrame(np.random.randint(1, 10, size=(4, 4)), columns=['A', 'B', 'C','D'])
print("Original DataFrame:")
print(df,'\n')

# Apply a function that doubles the values of the DataFrame
df_doubled = df.map(lambda x: x * 2)
print("Doubled DataFrame:")
print(df_doubled)

'''
df.map() is used to apply a function to **each individual cell** (element-wise operation) in a DataFrame.

In this example:
    lambda x: x * 2

• df.map()
  Applies this function to every value in the DataFrame.

Effect:
  Every element in `df` is doubled, producing a new DataFrame called `df_doubled`.

Note:
  df.map() works element-wise on the entire DataFrame.
  For column-wise operations, Pandas often uses `df.apply()` instead.
'''
```
