Write a Python function to create a table named `employees` with columns `id` (integer), `name` (text), `age` (integer), and `department` (text) in the `test.db` database.

```python
"""
SQLITE TABLE CREATION

This function creates a table named 'employees' in the SQLite database
'test.db' using Python's sqlite3 module.

Steps:

1. Connect to Database
   conn = sqlite3.connect('test.db')
   - Opens a connection to the database file.
   - If the database does not exist, it will be created automatically.

2. Create Cursor
   cursor = conn.cursor()
   - The cursor object allows execution of SQL commands.

3. Execute SQL Statement
   CREATE TABLE IF NOT EXISTS employees (...)
   - Creates a table called 'employees' if it does not already exist.

   Table structure:
       id          INTEGER PRIMARY KEY (unique identifier)
       name        TEXT NOT NULL (employee name, required)
       age         INTEGER (employee age)
       department  TEXT (department name)

4. Commit Changes
   conn.commit()
   - Saves the changes to the database.

5. Close Connection
   conn.close()
   - Releases database resources.

Result:
A table named 'employees' is created in test.db if it does not already exist.
"""

def create_table():
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS employees (
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL,
            age INTEGER,
            department TEXT
        )
    ''')
    conn.commit()
    conn.close()
    print("Table 'employees' created successfully.")

# Test the function
create_table()
```

### 2. Inserting Data

Write a Python function to insert a new employee into the `employees` table.

```python

"""
SQLITE DATA INSERTION

This function inserts a new employee record into the 'employees'
table in the SQLite database 'test.db'.

Steps:

1. Connect to Database
   conn = sqlite3.connect('test.db')
   - Opens a connection to the SQLite database.

2. Create Cursor
   cursor = conn.cursor()
   - Cursor is used to execute SQL commands.

3. Execute INSERT Query
   INSERT INTO employees (id, name, age, department)
   VALUES (?, ?, ?, ?)

   - Adds a new row to the employees table.
   - Uses parameterized placeholders (?) to safely insert values
     and prevent SQL injection.

4. Provide Values
   (id, name, age, department)
   - These values replace the placeholders in the SQL statement.

5. Commit Changes
   conn.commit()
   - Saves the inserted record to the database.

6. Close Connection
   conn.close()
   - Releases database resources.

Result:
A new employee record is added to the 'employees' table.

Example record inserted by the test call:
    id: 1
    name: Alice
    age: 30
    department: HR
"""

def insert_employee(id, name, age, department):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO employees (id, name, age, department)
        VALUES (?, ?, ?, ?)
    ''', (id, name, age, department))
    conn.commit()
    conn.close()
    print("Employee inserted successfully.")

# Test the function
insert_employee(1, 'Alice', 30, 'HR')
```

### 3. Querying Data

Write a Python function to fetch and display all records from the `employees` table.

```python
'''
FETCH ALL QUERY RESULTS

records = cursor.fetchall()

This retrieves all remaining rows from the result set of the most
recent SQL query executed by the cursor. 
'''

def fetch_all_employees():
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM employees')
    records = cursor.fetchall()
    conn.close()
    for record in records:
        print(record)

# Test the function
fetch_all_employees()
```

### 4. Updating Data

Write a Python function to update the department of an employee based on their `id`.

```python
"""
SQLITE UPDATE QUERY

1. UPDATE employees
   - Specifies the table where the change will occur.

2. SET department = ?
   - Updates the value of the 'department' column.
   - The ? is a placeholder for the new value.

3. WHERE id = ?
   - Ensures that only the employee with the specified ID is updated.
   - Prevents updating all rows in the table.

4. Parameterized Query
   (new_department, employee_id)
   - Safely inserts values into the placeholders.
   - Protects against SQL injection.
   - The order of values matches the order of placeholders.

"""

def update_employee_department(employee_id, new_department):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    cursor.execute('''
        UPDATE employees
        SET department = ?
        WHERE id = ?
    ''', (new_department, employee_id))
    conn.commit()
    conn.close()
    print("Employee department updated successfully.")

# Test the function
update_employee_department(1, 'Finance')
```

### 7. Handling Transactions

Write a Python function to insert multiple employees into the `employees` table in a single transaction. Ensure that if any insertion fails, none of the insertions are committed.

```python
"""
BATCH INSERT WITH TRANSACTION HANDLING (SQLite)

Function:
    insert_multiple_employees(employees)

Parameter:
    employees → list of tuples containing employee data
                (id, name, age, department)


Batch Insert with executemany()
   INSERT INTO employees (id, name, age, department)
   VALUES (?, ?, ?, ?)

   - executemany() inserts multiple rows efficiently.
   - Each tuple in the 'employees' list supplies values
     for one row in the table.

Transaction Handling
   try:
       cursor.executemany(...)
       conn.commit()

   - If all inserts succeed, changes are permanently saved.

Error Handling
   except Exception as e:
       conn.rollback()

   - If any insert fails (e.g., duplicate primary key),
     the entire transaction is rolled back.
   - No partial data is written to the database.

Purpose:
    Demonstrates safe batch insertion with transaction control
    to maintain database consistency.
"""


def insert_multiple_employees(employees):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    try:
        cursor.executemany('''
            INSERT INTO employees (id, name, age, department)
            VALUES (?, ?, ?, ?)
        ''', employees)
        conn.commit()
        print("All employees inserted successfully.")
    except Exception as e:
        conn.rollback()
        print("Error occurred, transaction rolled back.")
        print(e)
    finally:
        conn.close()

# Test the function with valid and invalid data
employees = [
    (6, 'Frank', 40, 'Finance'),
    (7, 'Grace', 29, 'Engineering'),
    (8, 'Hannah', 35, 'Marketing'),
    (9, 'Ivan', 38, 'Sales'),
    (6, 'Jack', 45, 'HR')  # Duplicate ID to cause an error
]
insert_multiple_employees(employees)
```

Write a Python function to update the age of multiple employees in a single transaction. Ensure that if any update fails, none of the updates are committed.

```python

"""
BATCH UPDATE OF EMPLOYEE AGES WITH TRANSACTION HANDLING (SQLite)

Function:
    update_multiple_employees_ages(updates)

Parameter:
    updates → list of tuples containing update data
              (new_age, employee_id)


3. Batch Update Using executemany()
   UPDATE employees
   SET age = ?
   WHERE id = ?

   - executemany() applies the UPDATE query to each tuple in the list.
   - Each tuple contains:
        (new_age, employee_id)

Example Input:
    updates = [
        (32, 1),
        (26, 2),
        (33, 3),
        (41, 4),
        (23, 5)
    ]

Behavior:
    - Each tuple updates the age of the employee with the specified ID.
    - If an ID does not exist, SQLite simply updates 0 rows and
      does NOT raise an error.
"""

def update_multiple_employees_ages(updates):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    try:
        cursor.executemany('''
            UPDATE employees
            SET age = ?
            WHERE id = ?
        ''', updates)
        conn.commit()
        print("All employee ages updated successfully.")
    except Exception as e:
        conn.rollback()
        print("Error occurred, transaction rolled back.")
        print(e)
    finally:
        conn.close()

# Test the function with valid and invalid data
updates = [
    (32, 1),
    (26, 2),
    (33, 3),
    (41, 4),  # Non-existing ID to cause an error
    (23, 5)
]
update_multiple_employees_ages(updates)
```

Modify the `employees` table to include a foreign key referencing the `id` column in the `departments` table, using executescript()

```python
"""
SQLITE SCHEMA MIGRATION USING executescript()

This code uses cursor.executescript() to execute multiple SQL statements
at once to modify the structure of the 'employees' table.

executescript() allows execution of an entire SQL script containing
multiple statements separated by semicolons.

Purpose:
Add a new column (department_id) with a foreign key reference to the
departments table while preserving existing employee data.

Steps:

1. Disable Foreign Key Checks
   PRAGMA foreign_keys = off;
   - Temporarily disables foreign key enforcement so table restructuring
     does not fail during migration.

2. Start Transaction
   BEGIN TRANSACTION;
   - Ensures the schema modification occurs atomically.

3. Rename Original Table
   ALTER TABLE employees RENAME TO old_employees;
   - Preserves existing data while freeing the table name.

4. Create New Table Structure
   CREATE TABLE employees (...)
   - Recreates the employees table with an additional column:
        department_id INTEGER
   - Adds a foreign key constraint referencing departments(id).

5. Copy Existing Data
   INSERT INTO employees (id, name, age, department)
   SELECT id, name, age, department FROM old_employees;
   - Moves existing records into the new table.
   - The new column department_id will initially be NULL.

6. Remove Temporary Table
   DROP TABLE old_employees;
   - Deletes the old table once migration is complete.

7. Commit Transaction
   COMMIT;
   - Saves all schema changes and data migration.

8. Re-enable Foreign Keys
   PRAGMA foreign_keys = on;
   - Restores enforcement of foreign key constraints.

Result:
The employees table is recreated with an additional column:

    department_id INTEGER
    FOREIGN KEY(department_id) REFERENCES departments(id)

while preserving all existing employee records.
"""

def add_department_foreign_key():
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    cursor.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;

    ALTER TABLE employees RENAME TO old_employees;

    CREATE TABLE employees (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER,
        department TEXT,
        department_id INTEGER,
        FOREIGN KEY(department_id) REFERENCES departments(id)
    );

    INSERT INTO employees (id, name, age, department)
    SELECT id, name, age, department FROM old_employees;

    DROP TABLE old_employees;

    COMMIT;

    PRAGMA foreign_keys=on;
    ''')
    conn.commit()
    conn.close()
    print("Table 'employees' modified successfully.")

# Test the function
add_department_foreign_key()
```
