### 4. Statistical Operations

Compute the mean, median, standard deviation, and variance of an array.

```python
array = np.random.randint(1, 21, size=(4, 4))
print(array)

# standard statistics, easily applied functions, builtin numpy
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)
variance = np.var(array)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
print("Variance:", variance)
```

Normalize a numpy array.

```python
# normalize -  scale the values to have a mean of 0 and a standard deviation of 1

array = np.arange(1, 10).reshape((3, 3))
print('array', array)

# Normalizing, can find this formula in statistics
mean = np.mean(array)
std_dev = np.std(array)
normalized_array = (array - mean) / std_dev

print("Normalized array:")
print(normalized_array)
```

Use boolean indexing to set all elements between 4 and 10 to 0 in an array.  
### V IMPORTANT

```python
# -------------------------------------------------------------
# NumPy Boolean Indexing Gotcha:
# 
# Example:
#   array[(array >= 5) & (array < 10)] = 0
#
# Explanation:
# 1. array >= 5  → returns a **boolean array** for each element
# 2. array < 10  → returns another **boolean array**
#
# ❌ Using 'and' between these arrays doesn't work:
#    - Python's 'and' expects **single boolean values**, not arrays
#    - Trying: array >= 5 and array < 10 
#      raises: ValueError: "The truth value of an array is ambiguous"
# 
# ✅ ✅ ✅ ✅ ✅  Correct approach: use '&' for **element-wise AND** ✅ ✅ ✅ ✅ ✅ ✅
#    - Parentheses are required: (array >= 5) & (array < 10)
#    - Without parentheses, operator precedence breaks it
#
# Why parentheses are required:
#    - '&' has **lower precedence** than comparison operators
#    - Writing array >= 5 & array < 10 would be evaluated as:
#          array >= (5 & array) < 10
#      which is incorrect
#
# Note: In normal Python 'if' statements:
#    - 'and' works on single booleans (True/False)
#    - No need for parentheses unless clarifying precedence
# -------------------------------------------------------------

array = np.arange(1, 16).reshape(3, 5)
print("Original array:")
print(array)

# simple conditional expression filters the array elements, and then assigning them new value
# and → works for single bool
# & → element-wise logical AND for arrays
array[(array >= 5) & (array <10) ] = 0
print("Modified array:")
print(array)
```

### 9. Structured Arrays

Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

### VERY IMPORTANT

```python
# -------------------------------------------------------------
# Structured NumPy Array Example
#
# 1. 'data_type' defines the **dtype for each field**:
#    - 'name'   : 'U10' → Unicode string of max length 10
#    - 'age'    : 'i4'  → 4-byte integer
#    - 'weight' : 'f4'  → 4-byte float
#
# 2. 'data' is a **structured array** where each element is like a row
#    with named fields (name, age, weight).
#
# 3. This allows you to access columns by name:
#       data['name']   → array(['Alice', 'Bob', 'Charlie'], dtype='<U10')
#       data['age']    → array([25, 30, 20], dtype=int32)
#       data['weight'] → array([55.5, 85.3, 65.2], dtype=float32)
#
# Structured arrays are like a hybrid of arrays and records:
# you can index like an array or access fields by name.
# -------------------------------------------------------------
data_type = [('name', 'U10'), ('age', 'i4'), ('weight', 'f4')]
data = np.array([('Alice', 25.9, 55.5), ('Bob', 30, 85.3), ('Charlie', 20, 65.2)], dtype=data_type)
print("Original array:")
print(data)

# Sort the array by age
sorted_data = np.sort(data, order='age')
print("Sorted array by age:")
print(sorted_data)
```

Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
# -------------------------------------------------------------
# Structured Array Example with Euclidean Distance
#
# 1️.  Create a structured array:
#    - 'data_type' defines fields: 'x' and 'y' as 4-byte integers
#    - 'data' stores each point as a tuple (x, y)
#
#       data = [(1,2), (3,4), (5,6)]
#       Access x-coordinates: data['x'] → [1 3 5]
#       Access y-coordinates: data['y'] → [2 4 6]
#
# 2️.   Compute Euclidean distance between all pairs of points:
#    - data['x'][:, np.newaxis] reshapes x-coords from (3,) → (3,1)
#      so broadcasting works for pairwise differences
#    - (data['x'][:, np.newaxis] - data['x'])**2 computes (x_i - x_j)^2 for all i,j
#    - Similarly for y: (data['y'][:, np.newaxis] - data['y'])**2
#    - np.sqrt(...) gives the Euclidean distance
#
# 3️.   Result:
#    distances[i,j] is the distance between point i and point j
#
# Example output:
# Original array:
# [(1, 2) (3, 4) (5, 6)]
#
# Euclidean distances:
# [[0.         2.828427  5.656854]
#  [2.828427  0.         2.828427]
#  [5.656854  2.828427  0.      ]]
#
# Notes:
# - Using structured arrays lets you access coordinates by name (data['x'], data['y'])
# - np.newaxis is used to **broadcast** arrays for pairwise operations
# -------------------------------------------------------------

data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)
print("Original array:")
print(data)

# Compute the Euclidean distance between each pair of points
distances = np.sqrt((data['x'][:, np.newaxis] - data['x'])**2 + (data['y'][:, np.newaxis] - data['y'])**2)
print("Euclidean distances:")
print(distances)
```

### 10. Masked Arrays

Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
# -------------------------------------------------------------
# NumPy Masked Arrays Example
#
# 1️⃣ What is a masked array?
#    - A masked array is a NumPy array that can **ignore certain elements** 
#      in computations without actually removing them.
#    - The "mask" marks which elements are **invalid or should be ignored**.
#    - This is useful when working with missing data, filtering, or ignoring outliers.
#
# 2️⃣ Creating the array:
#    - array = np.random.randint(1, 21, size=(4, 4)) → random integers 1 to 20
#    - masked_array = ma.masked_greater(array, 10)
#       - This **masks all elements greater than 10**
#       - Masked elements are ignored in computations like sum, mean, etc.
#
# 3️⃣ Example behavior:
#    Original array (random):
#    [[ 5 12  3 14]
#     [10  9 16  7]
#     [ 8 15  2 11]
#     [ 1 13  6  4]]
#
#    Masked array (elements >10 masked):
#    [[5 -- 3 --]
#     [10 9 -- 7]
#     [8 -- 2 --]
#     [1 -- 6 4]]
#    ('--' indicates masked/ignored elements)
#
# 4️⃣ Computations ignore masked elements:
#    - sum_unmasked = masked_array.sum() → sums only elements ≤10
#      (masked elements do not contribute)
#
# 5️⃣ Key Points:
#    - masked_array.mask → Boolean array showing which elements are masked
#    - masked_array.data → original underlying data
#    - Operations like sum(), mean() automatically **skip masked elements**
# -------------------------------------------------------------
import numpy.ma as ma

# Create a masked array of shape (4, 4) with random integers
array = np.random.randint(1, 21, size=(4, 4))
masked_array = ma.masked_greater(array, 10)
print("Original array:")
print(array)
print("Masked array:")
print(masked_array)

# Compute the sum of the unmasked elements
sum_unmasked = masked_array.sum()
print("Sum of unmasked elements:", sum_unmasked)
```

Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
# -------------------------------------------------------------
# NumPy Masked Array with Custom Mask
#
# 1️⃣ array: the original data array, e.g., shape (3,3) or larger
#
# 2️⃣ Mask: np.eye(3, dtype=bool)
#    - np.eye(3) creates a 3x3 identity matrix:
#          [[1, 0, 0],
#           [0, 1, 0],
#           [0, 0, 1]]
#    - dtype=bool converts 1 → True, 0 → False
#          [[True, False, False],
#           [False, True, False],
#           [False, False, True]]
#    - True values indicate **masked/ignored elements**
#      False values are kept.
#
# 3️⃣ Create masked array:
#    masked_array = ma.masked_array(array, mask=np.eye(3, dtype=bool))
#    - The **diagonal elements** of 'array' will be masked
#    - Off-diagonal elements remain unmasked
#
#
#    Computations:
#    - masked_array.sum() → sums only **unmasked elements**
#    - masked_array.mean() → ignores masked elements automatically
#
# 
#    You can create **any custom mask** using a Boolean array of the same shape
#    True → mask/ignore, False → keep
# -------------------------------------------------------------
array = np.random.randint(1, 21, size=(3, 3))
masked_array = ma.masked_array(array, mask=np.eye(3, dtype=bool))
print("Original array:")
print(array)
print("Masked array:")
print(masked_array)

# Replace the masked elements with the mean of the unmasked elements
mean_unmasked = masked_array.mean()
masked_array = masked_array.filled(mean_unmasked)
print("Modified masked array:")
print(masked_array)
```
